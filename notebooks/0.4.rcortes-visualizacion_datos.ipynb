```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion

x = np.arange( 2, 20)
y1 = x**2
y2 = x**0.5

plt.scatter(x , y1, color = 'green', marker='*', label='x^2')
plt.scatter(x , y2, color = 'red', marker='*', label='x^0.5')
plt.plot(x, y1, color='pink', label='line')

plt.title('Funciones Polin√≥micas' , fontsize=16, color = 'red')
plt.xlabel('x')
plt.ylabel('y1')
plt.ylabel('y2')
plt.grid()
plt.legend()

plt.text(5, 100, 'texto', color='blue')
```
