```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion

x = np.arange( 2, 20)
y1 = x**2
y2 = x**0.5

plt.scatter(x , y1, color = 'green', marker='*', label='x^2')
plt.scatter(x , y2, color = 'red', marker='*', label='x^0.5')
plt.plot(x, y1, color='pink', label='line')

plt.title('Funciones Polinómicas' , fontsize=16, color = 'red')
plt.xlabel('x')
plt.ylabel('y1')
plt.ylabel('y2')
plt.grid()
plt.legend()

plt.text(5, 100, 'texto', color='blue')
```

```python
# Histogramas

mean = 5
std = 3
n = 1000
x = np.random.normal(mean, std, n)
plt.hist(x, bins=20) ;

plt.title('Histograma de una variable aleatoria' , fontsize=16, color = 'red')
plt.xlabel('x')
plt.ylabel('frecuencia')
plt.grid()
plt.legend()

#plt.text(0, 80, f'Mean:{mean}, color='blue')
```

```python
plt.boxplot(x, vert=False);
```

```python
# Subplots
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(5, 10), sharex=True)
ax[0].hist(x);
ax[1].boxplot(x, vert=False);
```
